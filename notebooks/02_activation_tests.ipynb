# Part 2: Visualizing Activation Functions
The goal of this notebook is test `activation.py`. Visualizing **Sigmoid** and **ReLU** functions in two approachs:
* **Forward propagation:** Check how to funtions distort the input data $X$
* **Backward propagation:** Check the shape of Gradient $\frac{\partial A}{\partial Z}$ 

## Import files

```python
import numpy as np
import matplotlib.pyplot as plt
import sys

sys.path.append('../')
from core.activation import Sigmoid, ReLU
```

## 1. Prepare data

```python
X_test=np.linspace(-10,10,100).reshape(-1,1)
X_test
```

**Hypothetical error matrix**

```python
dummy_error=np.ones_like(X_test)
dummy_error
```

## 2. Testing and visualizing Sigmoid function 

### 2.1. Define and Run Layer

```python
sigmoid=Sigmoid()
Y_sig_forward=sigmoid.forward(X_test)
Y_sig_backward=sigmoid.backward(dummy_error,0.01)
```

### 2.2. Visualizing

```python
plt.figure(figsize=(10,5))
plt.plot(X_test,Y_sig_forward,label='Foward',color='blue',linewidth=2)
plt.plot(X_test,Y_sig_backward,label='Backward',color='red',linewidth=2,linestyle='--')
plt.title('Analysis Sigmoid Function',fontsize=14)
plt.xlabel('Input (Z)',fontsize=12)
plt.ylabel('Output/Gradient',fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```

## 3. Testing and visualizing ReLU function 

### 3.1. Define and Run Layer

```python
relu=ReLU()
Y_relu_forward=relu.forward(X_test)
Y_relu_backward=relu.backward(dummy_error,0.01)
```

### 3.2. Visulizing

```python
plt.figure(figsize=(10,5))
plt.plot(X_test,Y_relu_forward,label='Foward',color='blue',linewidth=2)
plt.plot(X_test,Y_relu_backward,label='Backward',color='red',linewidth=2,linestyle='--')
plt.title('Analysis ReLU Function',fontsize=14)
plt.xlabel('Input (Z)',fontsize=12)
plt.ylabel('Output/Gradient',fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```
