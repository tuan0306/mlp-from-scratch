# Part 4: Batch Inference and Kaggle Submission
Goals of this notebook are:
* **Load model:** Load best model from (`titanic_best_model.pkl`)
* **Batch Inference:** Put Kaggle test through the neuron network to predict survived or death.
* **Submission:** Export statistic report and save the `submission.csv` to submit to Kaggle.

## Import libraries

```python
import pickle
import numpy as np
import pandas as pd
import sys
sys.path.append('../')
from core.utils import prob_to_class
```

## 1. Load model and data

**Load model having been trained**

```python
with open('../models/titanic_best_model.pkl','rb') as f:
    model=pickle.load(f)
```

**Load Kaggle test**

```python
test_kaggle=np.load('../data/processed/test_kaggle.npy')
test_id_kaggle=np.load('../data/processed/test_id_kaggle.npy')
```

```python
test_kaggle
```

## 2. Batch Inference

```python
test_predictions_prob=model.predict(test_kaggle)
test_predictions_class=prob_to_class(test_predictions_prob).flatten()
```

```python
test_predictions_class.shape
```

## 3. Overall report

```python
survived_count=np.sum(test_predictions_class==1)
death_count=np.sum(test_predictions_class==0)

survived_rate=survived_count/len(test_predictions_class)*100
death_rate=death_count/len(test_predictions_class)*100

print(f'Number of passenger: {len(test_predictions_class)}')
print(f'Number of Survived: {survived_count} people ({survived_rate}%)')
print(f'Number of Death: {death_count} people ({death_rate}%)')
```

## 4. Export Submission

```python
submission_df = pd.DataFrame({
    'PassengerId': test_id_kaggle,
    'Survived': test_predictions_class
})

submission_path = '../submission.csv'
submission_df.to_csv(submission_path, index=False)
```
